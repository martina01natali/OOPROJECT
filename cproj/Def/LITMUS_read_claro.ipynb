```python
def linf(x,a,b):
    return a*x+b
```

```python
x_fit = single.x[2:5]
y_fit = single.y[2:5]
fit_params, fit_covar = curve_fit(linf, x_fit, y_fit)
```

```python
fit_params
```

```python
fit_dev = np.sqrt(np.diag(fit_covar))
fit_dev
```

```python
a = [fit_params[0], fit_dev[0]]
b = [fit_params[1], fit_dev[1]]
```

```python
x_grid = np.linspace(155,165, num=10)
y_grid = linf(x_grid, a[0], b[0])
```

```python
a_cpp = [145, 60.1]
b_cpp = [-22900, 9600]
y_cpp = linf(x_grid, a_cpp[0], b_cpp[0])
```

```python
plt.plot(single.x, single.y, marker='o', label='data')
plt.plot(x_grid,y_grid, label='python curvefit')
plt.plot(x_grid, y_cpp, label='cpp regression fit')
plt.legend()
```
