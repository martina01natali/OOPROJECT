```python
data = pd.read_excel("Correlazioni_Tommy_IA.xlsx","Dati", header=[0], engine="openpyxl")

K = data['K [%]']
Ar = data['Argilla [%]']

predSUP = (40.5710616958128+1.78336595582234) * K + (-12.0411376728981+1.67192004096413)
predINF = (40.5710616958128-1.78336595582234) * K + (-12.0411376728981-1.67192004096413)

predSUP2 = (40.5710616958128+(2*1.78336595582234)) * K + (-12.0411376728981+(2*1.67192004096413))
predINF2 = (40.5710616958128-(2*1.78336595582234)) * K + (-12.0411376728981-(2*1.67192004096413))

cont=0
cont2=0

for num in range(len(Ar)):
    if ((Ar[num]<predSUP[num]) and (Ar[num]>predINF[num])):
        cont+=1
    if ((Ar[num]<predSUP2[num]) and (Ar[num]>predINF2[num])):
        cont2+=1

print ('# 1sigma', '% 1sigma')
print (cont, str(round((cont/len(Ar))*100,0))+'%')
print()
print ('# 2sigma', '% 2sigma')
print (cont2, str(round((cont2/len(Ar))*100,0))+'%')
```
