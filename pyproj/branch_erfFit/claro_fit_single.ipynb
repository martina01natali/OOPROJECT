```python
# x,y data extraction and polishing

data = pd.read_csv('Test-file_Ch_0_offset_0_Chip_001.txt', sep='\t', header=None, skiprows=None)

AMPLITUDE = data.iloc[0,0]
TRANSITION = data.iloc[0,1]
WIDTH = -data.iloc[0,2]
GUESSES = [AMPLITUDE, TRANSITION, AMPLITUDE]

x = data.iloc[2:,0]
y = data.iloc[2:,1]
```

```python
# Fit the data with an erf function

def func(x, a, b, d):
    """
    The correct form of a modified erf function would be
    func(a, ampl, a, b)
        return ampl/2*(1+erf((x-a)/(b*np.sqrt(2))))
    """
    return a*sp.erf((x-b))+d

xgrid = np.linspace(x.min(), x.max(), 500)

popt, pcov = curve_fit(func, x, y, p0=GUESSES)
print(popt)
print(pcov)
```
